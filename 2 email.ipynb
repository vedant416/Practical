```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("./emails.csv")
df.head(20)
df.isnull().sum()
df.describe()

X = df.iloc[:,1:3001]
Y = df.iloc[:,-1].values
X
```

```python
# SVC
train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.25)

svc = SVC(gamma='auto')

svc.fit(train_x, train_y)       # train on traing data

prediction = svc.predict(test_x)  # give test to model

print("Accuracy Score for SVC : ", accuracy_score(prediction, test_y))  # test with predition
```

```python
# KNN
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)

prediction = knn.predict(X_test)

print("Accuracy Score for SVC : ", accuracy_score(prediction, y_test))  # test with predition
```
